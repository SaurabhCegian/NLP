```
import pandas as pd
df = pd.read_csv("/content/ner_dataset.csv", encoding = 'unicode_escape')

df.drop(["Sentence #"], axis = 1, inplace = True) 
df.head()
```

```
# find state vector
state_vector = []
for each_tag in df['Tag']:
  if(not each_tag in state_vector and  str(each_tag) != 'nan'):
    state_vector.append(each_tag)
state_vector
```

```
# find Start probability
start_probability = []
for each_vector in state_vector:
  len_particular_tag = len(df[df['Tag'] == each_vector])
  start_probability.append(len_particular_tag / df['Tag'].count())
start_probability
```

```
# find Transition probability
trans_prob_vect = []
def check_transition_probability(tagi, tagj):
  count = 0
  for i in range(len(df['Tag'])-1):
    if(df['Tag'][i] == tagi and df['Tag'][i+1] == tagj):
      count = count + 1
      i = i + 2
  trans_prob_vect.append(count)
  return trans_prob_vect

for i in range(len(state_vector)-1):
  tagi = state_vector[i]
  tagj = state_vector[i+1]
  trans_prob_vect = check_transition_probability(tagi, tagj)
print(trans_prob_vect)
```

```
# find emission probability
emission_prob = []
def find_emission_probability(word, tag):
  count = 0
  for i in range(len(df['Word']) - 1):
    if(word == df['Word'][i].lower() and df['Tag'][i] == tag):
      count = count + 1
  return (count / len(df[df['Tag'] == tag]))

for i in range(len(df['Word']) - 1):
  word = df['Word'][i]
  tag = df['Tag'][i]
  # print(word, tag)
  each_emission_prob = find_emission_probability(word.lower(), tag)
  emission_prob.append(each_emission_prob)
print(emission_prob)
```
