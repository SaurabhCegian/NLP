```
from nltk.util import bigrams, trigrams, ngrams
import nltk
nltk.download('punkt')
```

```
string = "Dummy text is text that is used in the publishing industry."
tokens = nltk.word_tokenize(string)
```

```
bi_tokens = list(nltk.bigrams(tokens)) #Bigrams
bi_tokens
```

```
tri_tokens = list(nltk.trigrams(tokens)) #trigrams
tri_tokens
```

```
n_tokens = list(nltk.ngrams(tokens, 9)) #Ngrams
n_tokens
```
